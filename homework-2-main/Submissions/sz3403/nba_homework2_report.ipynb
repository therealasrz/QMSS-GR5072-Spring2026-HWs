```python

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

nba = pd.read_csv(r'C:\Users\sirui\Downloads\Seasons_Stats_NBA.csv')

numeric_cols = ['Year', 'Age', 'PTS', 'FTA', 'FT%']
for col in numeric_cols:
    nba[col] = pd.to_numeric(nba[col], errors='coerce')

print("Shape:", nba.shape)
print("\nColumns:")
print(nba.columns.tolist())
```

## Q1. Which NBA player scored the most points in a single season? What season/year was this?

```python

q1 = nba.loc[[nba['PTS'].idxmax()], ['Player', 'Year', 'Tm', 'PTS']].reset_index(drop=True)

print("Answer:")
display(q1)

print("\nFirst five rows:")
display(q1.head())

print("\nLast five rows:")
display(q1.tail())
```

## Q2. Which NBA player scored the most points in 1991?

```python

nba_1991 = nba[nba['Year'] == 1991].copy()
q2 = nba_1991.sort_values(['PTS', 'Player'], ascending=[False, True]).head(1)[['Player', 'Year', 'Tm', 'PTS']].reset_index(drop=True)

print("Answer:")
display(q2)

print("\nFirst five rows:")
display(q2.head())

print("\nLast five rows:")
display(q2.tail())
```

## Q3. Which player had the best free throw percentage (FT%) from the year 2000 to the most recent year in the data? If there are ties, sort on player name and submit the answer as a DataFrame.

```python

nba_2000_plus = nba[nba['Year'] >= 2000].copy()

best_ft = nba_2000_plus['FT%'].max()
q3 = (
    nba_2000_plus[nba_2000_plus['FT%'] == best_ft][['Player', 'Year', 'Tm', 'FTA', 'FT%']]
    .sort_values('Player')
    .reset_index(drop=True)
)

print("Best FT% from 2000 onward:", best_ft)
print("Number of tied rows:", len(q3))

print("\nFirst five rows:")
display(q3.head())

print("\nLast five rows:")
display(q3.tail())
```

## Q4. Among those with at least 50 attempts (FTA), which player had the best free throw percentage (FT%) from the year 2000 to the most recent year in the data?

```python

nba_2000_plus_50fta = nba_2000_plus[nba_2000_plus['FTA'] >= 50].copy()

best_ft_50 = nba_2000_plus_50fta['FT%'].max()
q4 = (
    nba_2000_plus_50fta[nba_2000_plus_50fta['FT%'] == best_ft_50][['Player', 'Year', 'Tm', 'FTA', 'FT%']]
    .sort_values('Player')
    .reset_index(drop=True)
)

print("Best FT% from 2000 onward among players with at least 50 FTA:", best_ft_50)
display(q4)

print("\nFirst five rows:")
display(q4.head())

print("\nLast five rows:")
display(q4.tail())
```

## Q5. Rename the variable `Pos` to `Position`.

```python

nba_q5 = nba.rename(columns={'Pos': 'Position'}).copy()

print("Columns after renaming:")
print(nba_q5.columns.tolist())
```

## Q6. Use this variable to create two variables called `first_position` and `second_position`.

```python

position_split = nba_q5['Position'].str.split('-', n=1, expand=True)

nba_q6 = nba_q5.copy()
nba_q6['first_position'] = position_split[0]
nba_q6['second_position'] = position_split[1]

print("First five rows:")
display(nba_q6[['Player', 'Position', 'first_position', 'second_position']].head())

print("\nLast five rows:")
display(nba_q6[['Player', 'Position', 'first_position', 'second_position']].tail())
```

## Q7. Combine these two variables back into a single variable called `position_united`. Make sure to separate the two positions by `-`, as the original data field did. Present a frequency table of `position_united`.

```python

nba_q7 = nba_q6.copy()
nba_q7['position_united'] = np.where(
    nba_q7['second_position'].notna(),
    nba_q7['first_position'] + '-' + nba_q7['second_position'],
    nba_q7['first_position']
)

q7_freq = (
    nba_q7['position_united']
    .value_counts(dropna=False)
    .rename_axis('position_united')
    .reset_index(name='frequency')
)

print("First five rows:")
display(q7_freq.head())

print("\nLast five rows:")
display(q7_freq.tail())
```

## Q8. Create two new datasets.

```python

nba_without_age = nba.drop(columns=['Age']).copy()

nba_year_player_age = nba[['Year', 'Player', 'Age']].copy()

nba_without_age['mergeid'] = range(1, len(nba_without_age) + 1)
nba_year_player_age['mergeid'] = range(1, len(nba_year_player_age) + 1)

print("Dataset 1 (without Age) - first five rows:")
display(nba_without_age.head())

print("\nDataset 1 (without Age) - last five rows:")
display(nba_without_age.tail())

print("\nDataset 2 (Year, Player, Age) - first five rows:")
display(nba_year_player_age.head())

print("\nDataset 2 (Year, Player, Age) - last five rows:")
display(nba_year_player_age.tail())
```

## Q9. Join the two datasets from question 8 together to recreate the original dataset plus the new `mergeid`.

```python

q9_merged = pd.merge(
    nba_without_age,
    nba_year_player_age,
    on=['mergeid', 'Year', 'Player'],
    how='inner'
)

ordered_cols = ['mergeid', 'Year', 'Player', 'Age'] + [col for col in nba.columns if col not in ['Year', 'Player', 'Age']]
q9_merged = q9_merged[ordered_cols]

print("Shape of recreated dataset:", q9_merged.shape)

print("\nFirst five rows:")
display(q9_merged.head())

print("\nLast five rows:")
display(q9_merged.tail())
```

## Q10. Subset the original dataset to the years 1995-1997. Group the data by team name and year and then summarize the average number of points per team. Arrange from most to least points.

```python

nba_95_97 = nba[nba['Year'].between(1995, 1997)].copy()

q10 = (
    nba_95_97.groupby(['Tm', 'Year'], dropna=False)['PTS']
    .mean()
    .reset_index(name='avg_points')
    .sort_values('avg_points', ascending=False)
    .reset_index(drop=True)
)

print("Shape:", q10.shape)

print("\nFirst five rows:")
display(q10.head())

print("\nLast five rows:")
display(q10.tail())
```

## Q11. Create a wide dataset that keeps each year in its own row, but spreads team names to multiple columns with each column delineating points per team in each of the 1995-1997 seasons.

```python
q11 = (
    nba_95_97.groupby(['Year', 'Tm'], dropna=False)['PTS']
    .mean()
    .reset_index()
    .pivot(index='Year', columns='Tm', values='PTS')
    .sort_index()
)

print("Shape:", q11.shape)

print("\nWide dataset:")
display(q11)

print("\nFirst five rows:")
display(q11.head())

print("\nLast five rows:")
display(q11.tail())
```

## Q12. Return the data from Q11 back into a long (tidy) format by moving Team (TM), Points, and Year into their own columns.

```python

q12 = (
    q11.reset_index()
    .melt(id_vars='Year', var_name='Tm', value_name='PTS')
    .sort_values(['Year', 'Tm'])
    .reset_index(drop=True)
)

print("Shape:", q12.shape)

print("\nFirst five rows:")
display(q12.head())

print("\nLast five rows:")
display(q12.tail())
```

### Project Highlights

Metabase is popular because it is:

- Open source
- Designed for both technical and non-technical users
- Compatible with many databases such as MySQL, PostgreSQL, and BigQuery

Many companies use Metabase because it allows teams to query data in natural language and visualize results quickly.

## Features of Metabase <a class="anchor" id="bullet3"></a>

```python
import matplotlib.pyplot as plt

years = ["2019", "2020", "2021", "2022", "2023"]
companies_using_metabase = [15000, 30000, 50000, 70000, 90000]

plt.figure(figsize=(8, 5))
plt.plot(years, companies_using_metabase, marker="o")
plt.title("Growth of Companies Using Metabase")
plt.xlabel("Year")
plt.ylabel("Number of Companies")
plt.grid(True)
plt.show()
```
